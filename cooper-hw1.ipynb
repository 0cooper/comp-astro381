## Problem 2: Integral Processes

a) Numerically compute the integral

$\int_1^5 \frac{1}{x^{3/2}}dx$

with the rectangle and trapezoidal methods, and plot the error in the numerical integral against the step size
$\Delta x$ for both methods. Approximately how many steps are required to get an answer with
a fractional error $|I - I_{exact}|/I_{exact} < 10^{-3}$ for the rectangle and the trapezoid rule? What
about $10^{-5}$?

What did you lean about the trade-off between method, accuracy, and calculation speed?

```python
def int_proc(N,a=1,b=5,pwr=-3/2):
    """
    Computes integral using rectangle and trapezoid rules for f(x) = x^pwr
    
    INPUT:
        N = number of integration steps
        a = integral lower limit
        b = integral upper limit
        pwr = power of x for f(x) = x^pwr
        
    OUTPUT:
        np.array([delx,int_exact,int_rec,err_rec,frac_err_rec,int_trap,err_trap,frac_err_trap])
        
    """
    int_exact = (1/(pwr+1))*b**(pwr+1)-((1/(pwr+1))*a**(pwr+1)) # exact integral value
    delx = (b-a)/N # calc step size
    xarr = np.empty(shape=N) # init x array
    # assign x values
    for i in range(N):
        xarr[i] = a + (i-1)*delx
    # assign y values based on f(x array)
    yarr = xarr**(pwr)
    
    ### rectangle rule ###
    int_rec = np.sum(yarr*delx) # eval integral
    err_rec = np.abs(int_rec - int_exact) # eval error
    frac_err_rec = err_rec/int_exact # eval fractional error

    ### trapezoid rule ###
    int_trap = delx*np.sum([0.5*(yarr[j] + yarr[j+1]) for j in range(len(yarr)-1)]) # eval integral
    err_trap = np.abs(int_trap - int_exact) # eval error
    frac_err_trap = err_trap/int_exact # eval fractional error
    
    # dump results into array
    result = np.array([delx,int_exact,int_rec,err_rec,frac_err_rec,int_trap,err_trap,frac_err_trap]) 
    
    return result
```

```python
N = np.logspace(2,7,100,dtype=int) # init array of number of int steps, log spaced 10 - 10^5
int_tab = np.array([int_proc(n) for n in N]).T  # build table and transpose to be more easily indexed
```

```python
# plot step size versus error

plt.plot(int_tab[0],int_tab[3],'ms',label='Rectangle')
plt.plot(int_tab[0],int_tab[6],'cd',label='Trapezoid')
plt.xlabel(r'$\Delta x$')
plt.ylabel('Error')
plt.xlim(np.max(int_tab[0]),np.min(int_tab[0]))
plt.legend()
```

## Problem 3: Shooting for the stars

Write a program that computes the comoving distance,

$D_c = \int_0^z dz'[\Omega_m(1 + z')^3 + (1 - \Omega_m - \Omega_{\Lambda})(1 + z')^2 + \Omega_{\Lambda}]^{−1} $

```python
def comov(N,z,OmM=0.3,OmL=0.7):
    """
    Computes comoving distance integral from a given redshift using trapezoid rule
    
    INPUT:
        N = number of integration steps
        z = integral upper limit (redshift) 

    OUTPUT:
        
        dc = comoving distance in Mpc
    """

    delz = z/N # calc step size
    zarr = np.empty(shape=N) # init x array
    # assign x values
    for i in range(N):
        zarr[i] = (i-1)*delz
    # assign y values based on f(x array)
    darr = (OmM*(1+zarr)**3 + (1-OmM-OmL)*(1+zarr)**2 + OmL)**(-1/2)

    ### trapezoid rule ###
    int_trap = delz*np.sum([0.5*(darr[j] + darr[j+1]) for j in range(len(darr)-1)]) # eval integral
    
    # convert units into Mpc (times c/H0)
    dc = int_trap*3000
    
    return dc
```

```python
comov(N=1000,z=2)
```
