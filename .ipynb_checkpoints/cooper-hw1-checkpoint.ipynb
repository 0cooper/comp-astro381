## Problem 2: Integral Processes

a) Numerically compute the integral

$\int_1^5 \frac{1}{x^{3/2}}dx$

with the rectangle and trapezoidal methods, and plot the error in the numerical integral against the step size
$\Delta x$ for both methods. Approximately how many steps are required to get an answer with
a fractional error $|I - I_{exact}|/I_{exact} < 10^{-3}$ for the rectangle and the trapezoid rule? What
about $10^{-5}$?

What did you lean about the trade-off between method, accuracy, and calculation speed?

```python
def power(x,p):
    """
    Computes f(x) = x^pwr

    INPUT:
        x = x array
        p = power of x for f(x) = x^pwr

    OUTPUT:
        f(x) array

    """
    return x**(p)

def int_rect(N, a, b, pwr):   
    """
    Computes integral using rectangle rule for f(x) = x^pwr
    
    INPUT:
        N = number of integration steps
        a = integral lower limit
        b = integral upper limit
        pwr = power of x for f(x) = x^pwr
        
    OUTPUT:
        intg = integral computed from a to b
        
    """

    intg = 0
    delx = (b-a)/N
    for i in range(N):    
        xi = a+(i*delx)    
        intg = intg+(power(x=xi,p=pwr)*delx)
    return intg

def int_trap(N, a, b, pwr): 
    """
    Computes integral using trapezoid rule for f(x) = x^pwr
    
    INPUT:
        N = number of integration steps
        a = integral lower limit
        b = integral upper limit
        pwr = power of x for f(x) = x^pwr
        
    OUTPUT:
        intg = integral computed from a to b
        
    """
    intg = 0
    delx = (b-a)/N
    for i in range(N):    
        xi = a+(i*delx)    
        xi2 = a+((i+1)*delx)
        intg = intg+(delx*((power(x=xi,p=pwr)+power(x=xi2,p=pwr))/2.))
    return intg
```

```python
# evaluate integral
exact = -2/np.sqrt(5)+2   # analytical exact integral value

Narr = np.logspace(0,6,1000, dtype = int)

# estimate using rectangle rule
rec = [int_rect(n, a=1, b=5, pwr=(-3/2)) for n in Narr] # integral
frac_err_rec = (rec-exact)/exact # fractional error

# estimate using trapezoid rule
trap = [int_trap(n, a=1, b=5, pwr=(-3/2)) for n in Narr] # integral
frac_err_trap = (trap-exact)/exact # fractional error
```

```python
# plot results
step = 4/Narr    #stepsize = (b-a)/N
plt.loglog(step,frac_err_rec, 'ms',label='Rectangle')
plt.loglog(step,frac_err_trap, 'cd',label='Trapezoid')
plt.xlabel(r'$\Delta x$')
plt.ylabel('Error')
plt.legend()
```

```python
# compare fractional errors

err1 = 1e-3; err2 = 1e-5 # define error limits to test
err1_rec = Narr[closest(frac_err_rec,err1)[0]] # rectangle, 1e-3
err1_trap = Narr[closest(frac_err_trap,err1)[0]] # trapezoid, 1e-3
err2_rec = Narr[closest(frac_err_rec,err2)[0]] # rectangle, 1e-5
err2_trap = Narr[closest(frac_err_trap,err2)[0]] # trapezoid, 1e-5

print('A fractional error of 1e-3 requires:','\n',err1_rec,'steps for the rectangle method\n',\
      err1_trap,'steps for the trapezoid method')
print('\n')
print('A fractional error of 1e-5 requires:','\n',err2_rec,'steps for the rectangle method\n',\
      err2_trap,'steps for the trapezoid method')
```

There is a trade-off between method, accuracy, and calculation speed: for a given accuracy goal, we achieve a greater calculation efficiency when we improve the method rather than by increasing the number of steps. Here, using the trapezoid rule with fewer steps is more efficient than using the less accurate rectangle rule with more steps.

## Problem 3: Shooting for the stars

Write a program that computes the comoving distance,

$D_c = \int_0^z dz'[\Omega_m(1 + z')^3 + (1 - \Omega_m - \Omega_{\Lambda})(1 + z')^2 + \Omega_{\Lambda}]^{−1} $

```python
def comov(N,z,OmM=0.3,OmL=0.7):
    """
    Computes comoving distance integral from a given redshift using trapezoid rule
    
    INPUT:
        N = number of integration steps
        z = integral upper limit (redshift) 

    OUTPUT:
        
        dc = comoving distance in Mpc
    """

    delz = z/N # calc step size
    zarr = np.empty(shape=N) # init x array
    # assign x values
    for i in range(N):
        zarr[i] = (i-1)*delz
    # assign y values based on f(x array)
    darr = (OmM*(1+zarr)**3 + (1-OmM-OmL)*(1+zarr)**2 + OmL)**(-1/2)

    ### trapezoid rule ###
    int_trap = delz*np.sum([0.5*(darr[j] + darr[j+1]) for j in range(len(darr)-1)]) # eval integral
    
    # convert units into Mpc (times c/H0)
    dc = int_trap*3000/0.7
    
    return dc
```

```python
comov(N=1000,z=2)
```

The comoving distance to $z=2$ in a universe with $\Omega_m=0.3$ and $\Omega_{\Lambda}=0.7$ is $\sim5186$ Mpc. Using Ned Wright's cosmology calculator with these parameters, I find the comoving distance to be $5179.2$ Mpc, which is approximately consistent.

```python
# plot comoving distance [Mpc] versus redshift for z = 0 − 10
zarray = np.arange(0,10,0.01)
comdist = np.array([comov(N=100,z=z) for z in zarray])
plt.plot(zarray,comdist,'c-')
plt.xlabel(r'$z$')
plt.xlim(0,10)
plt.ylim(0,1e4)
plt.ylabel(r'$D_c$ [Mpc]')
plt.show()
```
